There are several fields of text in this dataset. For now let us isolate the helpdesk "answer" field. We will first be looking at word count.

## Word count

```python
def analyze_word_count(df, column='answer'):
    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in DataFrame")

    # Calculate word count for each row in the specified text column
    df['word_count'] = df[column].apply(lambda x: len(str(x).split()))

    return df
```

```python
analyze_word_count(df_english)
```

```python
def analyze_word_count_and_category(df, count_column='word_count', category_column='priority'):
    word_count_by_category = df.groupby(category_column)['word_count'].mean().reset_index()
    
    plt.figure(figsize=(8, 6))
    sns.barplot(data=word_count_by_category, x=category_column, y=count_column)
    plt.title('Average Word Count by Category')
    plt.xlabel('Category')
    plt.ylabel('Average Word Count')
    plt.show()
```

```python
analyze_word_count_and_category(df_english)
```

Ok. Do not observe correlation between priority of customer ticket and the length of the helpdesk answer.

## TF-IDF
This highlights words that are important in a document, relative to the rest of the dataset.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def apply_tfidf(df, text_column='answer'):
    vectorizer = TfidfVectorizer(stop_words='english') # removes english stopwords
    
    tfidf_matrix = vectorizer.fit_transform(df[text_column])
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names())
    
    return tfidf_df
```

```python
tfidf_df = apply_tfidf(df_english)
tfidf_df
```

### TF-IDF mean

```python
def print_mean_tfidf(tfidf_df):
    mean_tfidf = tfidf_df.mean().sort_values(ascending=False)
    print("Top terms by mean TF-IDF:")
    print(mean_tfidf.head(10))
```

```python
print_mean_tfidf(tfidf_df)
```

### Top n terms of each document by TF-IDF

```python
def get_top_n_terms_per_doc(tfidf_df, n=5):
    top_terms_per_doc = tfidf_df.apply(lambda row: row.nlargest(n).index.tolist(), axis=1)
    print("Top terms for each document:")
    print(top_terms_per_doc)
    return top_terms_per_doc
```

```python
get_top_n_terms_per_doc(tfidf_df)
```

### TF-IDF k-means clustering
Group documents based on similar term distributions.

```python
from sklearn.cluster import KMeans

def get_tf_idf_kmeans_clusters(tfidf_df, df, n_clusters=5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(tfidf_df)
    
    # Assign cluster labels to each document, to original dataframe df
    df['tfidf_cluster'] = kmeans.labels_
    return df
```

```python
get_tf_idf_kmeans_clusters(tfidf_df, df_english)
```

```python
def cluster_vs_category(df, cluster_column='tfidf_cluster', category_column='priority'):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=cluster_column, hue=category_column)
    plt.title(f"Distribution of '{category_column}' within each Cluster")
    plt.xlabel("Cluster")
    plt.ylabel("Count")
    plt.legend(title=category_column)
    plt.show()
```

```python
cluster_vs_category(df_english)
```

#### TF-IDF clusters and chi-square test of independence
Statistical test to determine if there is a significant association between two categorical variables, in this case the tf-idf cluster a document has been assigned, and its priority (could replace with a different categorical field).

Null hypothesis: two categorical variables are independent.

"A p-value less than 0.05 is typically considered to be statistically significant, in which case the null hypothesis should be rejected."

```python
from scipy.stats import chi2_contingency

def chi_test_cluster_vs_category(df, cluster_column='tfidf_cluster', category_column='priority'):
    # create contingency table between cluster labels and a categorical field (e.g., "priority")
    contingency_table = pd.crosstab(df[cluster_column], df[category_column])
    # Perform chi-square test
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    print(f"Chi-square statistic: {chi2}, p-value: {p}")
```

```python
chi_test_cluster_vs_category(df_english)
```

## Latent Dirichlet Allocation (LDA) topic modeling

```python
from nltk.corpus import stopwords
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    # tokenize and remove stopwords
    tokens = [word for word in text.lower().split() if word.isalpha() and word not in stop_words]
    return tokens
```

```python
def get_dominant_topics(lda_model, bow, threshold=0.1):
    # Get topic distribution for the document
    topic_probs = lda_model.get_document_topics(bow)
    # Filter topics based on a probability threshold
    # If probability above threshold, topic is relevant for document
    dominant_topics = [topic for topic, prob in topic_probs if prob >= threshold]
    return dominant_topics

def LDA(df, field='answer', resultfield='processed_answer', n_topics=5):
    # Preprocess
    df[resultfield] = df[field].apply(preprocess_text)
    
    # Create bag-of-words
    dictionary = Dictionary(df['processed_answer'])
    corpus = [dictionary.doc2bow(text) for text in df['processed_answer']]
    
    # Train LDA model
    lda_model = LdaModel(corpus=corpus, num_topics=n_topics, id2word=dictionary, passes=10, random_state=0)
    print("Topics:")
    for idx, topic in lda_model.print_topics(-1):
        print(f"Topic {idx + 1}: {topic}") # print topics, i.e. most characteristic terms of each topic
        
    # Assign topics to each document
    df['assigned_topics'] = [get_dominant_topics(lda_model, bow) for bow in corpus]
```

```python
LDA(df_english)
```
